# Control Structures / Flow Control

Control structures decide the flow of the program. Python supports three control structures:

1. Sequence
2. Selection/Decision (also called, Conditional Statements)
3. Iteration (also called, Loops)

Let's look at them one by one.

## 1. Sequence

Sequential statements are a set of statements whose execution process happens in a sequence. We have been using it till now.

The problem with sequential statements is that if the logic has broken in any one of the lines, then the complete source code execution will break.

```python
# We will initialize some variables
# Then operations will be done
# And, at last, results will be printed
# Execution flow will be the same as the code is written, and there is no hidden flow

a = 20
b = 10
c = a - b
d = a + b
e = a * b
print("The result of the subtraction is: ", c)
print("The result of the addition is: ", d)
print("The result of the multiplication is: ", e)
```

## 2. Selection

These are also called **Branching**, **Conditional** or **Decision Control** Statements.

Output of indented block appear ONLY IF the given condition before it is True.

<img src = "https://miro.medium.com/v2/resize:fit:916/1*TSHmlVfBaO2vTtvOL8E43Q.png" width = 400>

**IMPORTANT**: Role of **indentations** is very much important in conditional statements and in many more topics you'll learn moving forward.

```python
# Single condition

score = 90

if score >= 85:   # next code block will work ONLY IF the condition in this line is True
    grade = 'A'
    print("The grade is:", grade)

print("Program has ended.")

# try changing score = 70 and see what happens. Why?
```

```python
# Two conditions

score = 80

if score >= 85:
    grade = 'A'

else:     # in all other conditions (otherwise)
    grade = 'B'

print("The grade is:", grade)
```

```python
# Multiple conditions

score = 60

if score >= 85:
    grade = 'A'

elif score >= 70:   # elif = else + if
    grade = 'B'

elif score >= 55:
    grade = 'C'

else:     # in all other conditions (otherwise)
    grade = 'Fail'


print("The grade is:", grade)
```

## 2. Iteration/Loops

Loops help us repeat tasks.

There are generally two loop statements to implement the repetition structure:

1. `while` loop
2. `for` loop

### 2.1 `while` loops

**Keep doing something as long as the given condition is True**

A `while` loop when looping is based on a conditional that will be modified during the loop. The loop continues to work until a given precondition is satisfied. Once the condition becomes `False`, the `while` loop stops.

The number of repeatitions is **not known** in a `while` loop. It continues AS LONG AS the given condition is `True`.

A `while` loop is pretty simple, it's structure looks like:

    while a_condition:
        # do something
        ...
        
The most important part to always keep track of is that you finish the condition that you started the `while` loop with. Otherwise, it'll just keep going on and on for forever!

**Flow chart of a `while` loop:**

<img src = "https://pynative.com/wp-content/uploads/2021/06/while-loop-flowchart.png" width = 250>

```python
# Print all numbers from 0 to 5 using WHILE loop

x = 0

while x <= 5:  # this is a condition which results in True/False
    print(x)   # this is part of code block which will run ONLY IF or TILL THE TIME the given condition in True
    x = x + 1

print("The loop has ended.")  # This is outside the loop hence, it will work ONLY after the loop has ended.
```

**Note**:
- If you do not give an increment of the variable x, the condition ( 0<= 5) will always be True.
- If you have written an infinite loop, please go to Kernel -> Interrupt and then give some increment.

### 2.2  `for` loop

Iterating through a sequence

In `for` loop, number of iterations are known.

A `while` loop is an excellent choice when you need to perform a set of commands and do not know how many times they **should** be executed but do know when the commands **should finish**.

But, with `for` loop, we directly or indirectly mention **how many times** the command or task should be repeated.


A `for` loop lets us repeat a set of commands a defined number of times. The syntax for a `for` loop is just:

    for item in sequence:
        # do something with item
        ...

**Flow chart of a `for` loop**

<img src = "https://pynative.com/wp-content/uploads/2021/06/for-loop-flowchart.png" width = 350>


But what is a sequence?

There are lots of functions in Python that will actually return a sequence - they are called *iterators*. An iterator essentially provides the next element in the sequence each time we access it.

The iterator that we will use to demonstrate a for loop is the `range()` function. The range function gives us a sequence of numbers from the first number we give it up until the last number we give it.

```python
# Use FOR loop to print all numbers whole number till 4

for i in range(5):   # range 5 means numbers [0,1,2,3,4]
    print(i)  # In each loop, the value of i will change from one item to the next item


# We are indirectly saying that run the loop to the number of times equal to the number of items in the list
```

```python
# Use FOR loop to print range of numbers with lower limit and upper limit mentioned.

for i in range(1, 6):  # upper range is excluded
    print(i)

print("Loops has ended.")
```

```python
# Since FOR uses a sequence, we can use a list or tuple of values as well

fruits = ['banana', 'apple', 'grapes']

for fruit in fruits:
    print(fruit)
```

```python
# The sequence can work on a string as well as it is a sequence of characters

word = 'Python'
for i in word:
    print(i)
```

What's happening is that each time we go to the top of the for loop, we pull another number out of the sequence and that number is assigned to `i`.

After that point, we execute all of the indented code in the block with the current value of `i`. Once we've finished executing the code we go back up to the top and assign the next value in the sequence to `i`.
