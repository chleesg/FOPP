# 1. `random` library

The Python Random module is a built-in module for generating random integers in Python. These are sort of fake random numbers which do not possess true randomness. We can therefore use this module to generate random numbers, display a random item for a list or string, and so on.

Let's see some examples of `random` functions.

```python
# First, we need to import the library

!pip list
```

Now, to use any function of `random` library, we need to use random.FunctionName()

Later, we will see how to use alias names for libraries.

```python
# using Random library to create OTPs of 6 digits

OTP = random.randint(100000, 999999)  # both values are included

# randint() generates a random integer from a range of values.
```

```python
# Print a random item from a sequence like a list

lottery_numbers = [456234, 897654, 568123, 340876, 123876, 569234, 719284]

winner= random.choice(lottery_numbers)

print(winner)
```

```python
# Return a random sample from a big data

mylist = ["John", "Lily", "Robert", "Kevin", "Nastya", "Rebecca", "Leah", "Akbar", "Patel"]

print(random.sample(mylist, k=2)) # k indicating how many items to print in the sample
```

`numpy` offers `array` data structures which are similar to list. The difference is that lists do not calculations while arrays do. Hence, use of `numpy` library becomes important.

```python
NumberList = [4,5,6,7]
print(NumberList*2)  # This will just repeat the numbers as the input is a list.

NumberArray = numpy.array([4,5,6,7])
print(NumberArray*2)   # This will multiply each number with 2 as the numbers are from an array.
```

# 4. `matplotlib` library

Data visualization is a very important part of data analysis. You can use it to explore your data. If you understand your data well, you’ll have a better chance to find some insights.

There are many libraries in Python for data visualization. You will explore 3-4 different libraries in the next course. Here, we will look into just a few functions of the most commonly used library - `matplotlib` and its module `pyplot`


```python
import matplotlib.pyplot as plt # plt is the alias name - makes life easier
```

```python
# LINE PLOT

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
unemployment_rate = [8, 7.2, 6.9, 7, 6.5, 6.2, 9.5, 7.3]

plt.plot(years, unemployment_rate)
```

There are many functions in `matplotlib.pyplot` which can be used to customize a chart or plot area. We will explore some of those.

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
unemployment_rate = [8, 7.2, 6.9, 7, 6.5, 6.2, 9.5, 7.3]

plt.plot(years, unemployment_rate, c = "brown") # c for color

plt.title('Unemployment rate vs year')
plt.xlabel('Year')
plt.ylabel('Unemployment rate')

plt.show()  # Try to comment this line and see how output appears.
# plt.show() is just to print the plot area only, not how the object appears at backend
```

Next, we will create a bar chart.

```python
# BAR PLOT

names = ["John", "Lily", "Robert", "Kevin", "Nastya"]
scores = [8, 7, 5, 6, 5]

plt.bar(years, unemployment_rate)

# Use plt.barh() in the previous line for horizontal bar-chart
# Please note the first record will show up at last in .barh()

plt.title('Exam Result')
plt.xlabel('Student Names')
plt.ylabel('Scores')

plt.show()
```

To understand the coorelation between different variables/columns, we can create a scatter plot.

```python
# SCATTER PLOT to understand the relationship between monthly Income vs Expense

income = [6000, 4000, 4400, 5000, 3800, 5500]
expense = [4800, 3500, 3500, 4000, 3000, 4000]

# Change size of figure
plt.figure(figsize=(5,3))  # setting up the chart area/size

plt.scatter(income, expense)  # for scatter plot, both axis should have numbers

plt.title("Income and Expense relationship of 6 households")
plt.xlabel("Income")
plt.ylabel("Expense")
plt.show()
```

### Subplots using matplotlib - When multiple plots need to be shown

```python
charts = plt.figure(figsize = (15,5))  # We are defining this as a variable so that later we can save it

ax1 = plt.subplot(1,3,1)  # rows, columns, panel number
ax2 = plt.subplot(1,3,2)
ax3 = plt.subplot(1,3,3)

ax1.plot(years, unemployment_rate)
ax1.title.set_text('Unemployment rate vs year')

ax2.bar(names, scores)
ax2.title.set_text('Exam Result')

ax3.scatter(income, expense)
ax3.title.set_text("Income and Expense relationship of 6 households")

plt.show()
```

### How do you save figures to a file?

You can either right-click on the above plotting output and click on "Save the image as" or, can use `savefig()` to export the image.

```python
# Saving the 'charts' plot area in our directory

charts.savefig('charts')

# Now check your directory and you'll find an image file there.
```
